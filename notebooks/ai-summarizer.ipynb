````python
import os
import json
from dotenv import load_dotenv
from openai import OpenAI
from textwrap import dedent
````

````python
client = OpenAI(api_key=os.getenv('OPENAI_API_KEY'))
````

````python
# load AMZN_news.json file
with open('../data/AMZN_news.json') as f:
    articles = json.load(f)

print(articles)
````

````python
def summarize_article(article: dict, ticker: str):
    model = 'gpt-4o'
    system_message = """
    You are a financial news summarizer.
    Your job is to summarise financial news articles in a few sentences and give actionable yet balanced insights.
    Output the summaries in a clear and concise manner.
    The summaries should almost sound like tweets to be engaging and informative.
    Each summary should be around 280 characters long.
    Add emojis but no hashtags.
    """
    user_message = f"""
    Summarize the following article:
    ```
    Title: {article['title']}
    Content: {article['content']}
    ```
    Note that the article may contain irrelevant information, only summarise the information related to '{ticker}'.
    """
    system_message = dedent(system_message).strip()
    user_message = dedent(user_message).strip()
    completion = client.chat.completions.create(
        model=model,
        messages=[
            {'role': 'system', 'content': system_message},
            {'role': 'user', 'content': user_message}
        ]
    )
    return completion.choices[0].message.content

article = articles[0]
completion = summarize_article(article, ticker='AMZN')
print(completion)
````
